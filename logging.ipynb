## Logging in Python
Logging in Python is used to record messages about what a program is doing.
It helps you track errors, warnings, and program activity.

Instead of using print(), professionals use the logging module.

Logging is a crucial aspect of any application , providing a way to track events, errors, and operational information. Python's built-in logging module offers a flexible framework for emitting log messages from Python programs. 

```python
import logging

## configure the basic logging settings
logging.basicConfig(level=logging.DEBUG) # Configure logging and show all log messages including debugging details.

## log messages with different different severity levels
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning messsage")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

Python's logging module has several log levels indicating the severity of the events. The default levels are :- 

DEBUG → Detailed information used for debugging the program.

INFO → General information about normal program execution.

WARNING → Indicates something unexpected but the program still works.

ERROR → Shows that a problem occurred and a task failed.

CRITICAL → Indicates a very serious error that may stop the program.

```python
import logging

logging.basicConfig(
    filename="app.log",
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

logging.info("Application started")
logging.error("Database connection failed")
# this logs and their time stamps are stored in the log file
## log messages with different different severity levels
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning messsage")
logging.error("This is an error message")
logging.critical("This is a critical message")
```
