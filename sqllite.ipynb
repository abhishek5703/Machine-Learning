## SQL and SQLite
SQL (Structured Query Language) is a language used to communicate with databases. On the other hand SQLite is a lightweight database system that uses SQL.It is a database engine It does not need a server Everything is stored in one file (.db)

```python
import sqlite3
```

```python
## Connect to an SQLite database
connection = sqlite3.connect('example.db')  # this line will automatically create an example.db file since SQLite database works and stores locally and it can be created and stored in our local system without any setup
connection
```

```python
# When using SQLite in Python, the cursor is used to execute SQL commands and fetch results from the database.
# Connection  → connects Python to the database
# Cursor  → sends SQL commands and reads results
cursor = connection.cursor()
```

```python
# Create a Table
cursor.execute("""
    CREATE TABLE employees (    
    id integer primary key,
    name text Not Null,
    age integer,
    department text
)
""")

## Commit the changes
connection.commit()  # table has been created 
```

```python
# it will return me an object as a response
cursor.execute("""
    select * from employees
""")
```

```python
# Insert some data in the SQLite Table
## 1st record inserting 
cursor.execute('''
    insert into employees(name,age, department)
               values("Abhishek",22,"ML")
''')

## 2nd record inserting 
cursor.execute('''
    insert into employees(name,age, department)
               values("Bob",23,"CL")
''')

## 3rd record inserting 
cursor.execute('''
    insert into employees(name,age, department)
               values("ASh",25,"DL")
''')

## Commit the changes 
connection.commit()
```

```python
## Quering the data from the table
cursor.execute('select * from employees')
rows = cursor.fetchall() # fetch all the rows of the table 

## print the queried data ( The rows of the table)
for row in rows:
    print(row)
```

```python
## update the data in the table
cursor.execute('''
    update employees
    set age = 34
    where name = "ASh"
''')

# here the name is the primary key
connection.commit()
```

```python
## Quering the data from the table
cursor.execute('select * from employees')
rows = cursor.fetchall() # fetch all the rows of the table 

## print the queried data ( The rows of the table)
for row in rows:
    print(row)
```

```python
## Deleting the data from the table
cursor.execute('''
    delete from employees
    where name = "ASh"
''')
connection.commit()
```

```python
## Quering the data from the table
cursor.execute('select * from employees')
rows = cursor.fetchall() # fetch all the rows of the table 

## print the queried data ( The rows of the table)
for row in rows:
    print(row)
```

```python
## Working with a sales data
# Connect to a SQLite database

connection = sqlite3.connect('sales_data.db')
cursor = connection.cursor()

# create a table for the sales data
cursor.execute('''
    CREATE TABLE IF NOT EXISTS sales(
    id  INT PRIMARY KEY,
    date TEXT NOT NULL,
    product TEXT NOT NULL,
    sales INTEGER,
    region TEXT      
    )
''')

# Insert the data into the tables
sales_data = [
    (1, "2024-01-01", "Laptop", 1200, "North"),
    (2, "2024-01-02", "Mobile", 800, "South"),
    (3, "2024-01-03", "Tablet", 600, "East"),
    (4, "2024-01-04", "Headphones", 300, "West"),
    (5, "2024-01-05", "Smartwatch", 450, "North"),
    (6, "2024-01-06", "Camera", 900, "South")
]

# many rows are inserted into the table at once
cursor.executemany('''
    INSERT INTO sales (id, date, product, sales, region) VALUES (?, ?, ?, ?, ?)
''',sales_data)

connection.commit()
```

```python
## Quering the data from the table
cursor.execute('select * from sales')
rows = cursor.fetchall() # fetch all the rows of the table 

## print the queried data ( The rows of the table)
for row in rows:
    print(row)
```

```python
# close the connection
connection.close()
```
