## Pandas
Pandas is a powerful open-source Python library used for data manipulation and analysis. It allows users to easily work with structured data such as tables, similar to Excel or SQL databases. Pandas provides two main data structures: Series (1-dimensional data) and DataFrame (2-dimensional tabular data). With Pandas, you can read data from files like CSV or Excel, clean and preprocess datasets, filter and transform data, handle missing values, and perform statistical analysis. It is widely used in data analytics, data science, and machine learning, often together with libraries like NumPy for numerical operations and scikit-learn for building machine learning models.

## Series
In pandas, a Series is a one-dimensional labeled array that can store data of any type (integers, strings, floats, etc.). It is similar to a single column in a table or an Excel sheet, where each value has an index (label) and a corresponding data value. A Series helps in storing and manipulating a sequence of values efficiently and is built on top of NumPy arrays.

```python
#Creating a Simple Series
import pandas as pd

data = pd.Series([10, 20, 30, 40])
print(data)
# Here 0,1,2,3 are indexes and 10,20,30,40 are values.
```

```python
# Series with Custom Index
import pandas as pd

data = pd.Series([90, 85, 88], index=["Math", "Science", "English"])
print(data)

# You can access values using the index:
print(data["Math"])  # it will print 90 since the "math" index stores 90
```

```python
import pandas as pd
data = [10,20,30] # data of the series
index = ['a','b','c'] # indexes for the data 
pd.Series(data , index = index)
```

```python
# Series from Dictionary
import pandas as pd

data = {"A": 10, "B": 20, "C": 30}
s = pd.Series(data)

print(s)
```

NOTE:- In short: A Series = Index + Values, representing one column of data in Pandas.

## DataFrame 
In pandas, a DataFrame is a two-dimensional labeled data structure that stores data in rows and columns, similar to a table in Excel or a database. Each column can contain different types of data such as numbers, strings, or dates. It is the most commonly used structure in Pandas for analyzing datasets. Internally, a DataFrame is built using NumPy arrays.

```python
# Creating a DataFrame
import pandas as pd

data = {
    "Name": ["Abhishek", "Rahul", "Aman"],
    "Age": [21, 22, 23],
    "City": ["Delhi", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Accessing Data
df["Name"]      # access a column
df.iloc[0]      # access first row
df.head()       # first 5 rows
```

```python
# converting the df(data frame) into a numpy array 
import numpy as np
np.array(df)
```

```python
# In pandas, you can create a DataFrame from a list of dictionaries, where each dictionary represents one row and the keys become column names.
import pandas as pd

data = [
    {"Name": "Abhishek", "Age": 21, "City": "Delhi"},
    {"Name": "Rahul", "Age": 22, "City": "Mumbai"},
    {"Name": "Aman", "Age": 23, "City": "Pune"}
]

df = pd.DataFrame(data)

print(df)
```

```python
df.drop("Grade", axis=1, inplace=True)
# The axis parameter tells Pandas whether you want to drop a row or a column. axis=0 → drop rows or axis=1 → drop columns and inplace=True -> this tells Pandas to modify the original DataFrame directly. True → changes the original df and False → returns a new DataFrame without changing the original
```
