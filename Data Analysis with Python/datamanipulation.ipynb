pandas is widely used for data manipulation and data analysis in Python. Data manipulation means modifying or transforming the dataset (adding, removing, filtering, or updating data), while data analysis means extracting useful insights such as averages, counts, or patterns from the data.

# handling missing values

```python
# Handling Missing Values
df.isnull()  # Check where values are missing and gives true for a missing value
```

```python
df.isnull().sum() # Count missing values in each column
```

```python
df.isnull().any(axis=1) # Check each row(record) and If any column has missing value → True
```

```python
df.isnull().any() # here it takes bydefault axis = 0 and check each column wise 
```

```python
df.fillna(0) # replace all the missing values with 0
# usually we make an another copy of the data and then apply the changes like this
df_filled = df.fillna(0)  # make copy so that the original data should never be hampered
# generally we don't fill the missing values with the 0s 
```

```python
# filling the missing values with the mean of the column
df['Sales'].fillna(df['Sales'].mean()) # filling the missing values of the sales column with the mean of the sales column

df['Sales_fillNA'] = df['Sales'].fillna(df['Sales'].mean()) #Creates a new column called Sales_fillNA.This column contains the Sales values but with missing values replaced by the mean. So the original column stays unchanged.
```

```python
# Rename the column
df = df.rename(columns={'Date' : 'Sale Date'})
df.head()
```

```python
# Change the data types
df['New_Vlaues'] = df['Value'].astype(float)  # a new column of float values are created 
df.head()
```

```python
# apply a function to a column
# df['column_name'].apply(function)
df['New Value'] = df['Value'].apply(lambda x : x*2) # applied the lambda function to the col
df.head()
```

```python
# Data Aggregation and Grouping
grouped_mean = df.groupby('Product')['Value'].mean() # This groups the dataframe by the Product column and then for each product type show the mean of the value 
print(grouped_mean)
```

```python
grouped_data = df.groupby(['Product','Region'])['Value'].sum() # grouping of the data by using 2 columns and then get the sum of value columns
print(grouped_data)
```

```python
grouped_data = df.groupby(['Product','Region'])['Value'].mean()
print(grouped_data)
```

```python
# Multiple aggregation functions , applying multiple aggregate functions to a column or colums
grouped_agg = df.groupby('Region')['Value'].agg(['mean','sum','count'])
print(grouped_agg)
```

## Merging and joining DataFrames

In Pandas, merging and joining are used to combine multiple DataFrames into one.

This is very common in data analysis and ML preprocessing.

Think of it like joining tables in SQL.

merge() combines DataFrames using a common column (key).

```python
# Creating sample data frames and here the key and the values are the columns
df1 = pd.DataFrame({'key' : ['A','B','C'] , 'Values1' : [1,2,3]}) 
df2 = pd.DataFrame({'key' : ['A','B','D'] , 'Values2' : [4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merge this 2 data frames into the 'key column'
pd.merge(df1,df2,on="key",how="inner")  # In Pandas, Inner Join keeps only the rows that exist in both DataFrames based on the key column. So it returns the common matching rows.
```

```python
pd.merge(df1, df2, on="key", how="outer") 
# In Pandas, Outer Join keeps all rows from both DataFrames.
# If a match is not found, the missing values are filled with NaN.
```

```python
pd.merge(df1, df2, on="key", how="left")
# In Pandas, a Left Join keeps all rows from the left DataFrame and only the matching rows from the right DataFrame.
# If there is no match, the missing values become NaN.
# df1 → left table
# df2 → right table
# on="key" → column used for matching
# how="left" → keep all rows from df1
```

```python
pd.merge(df1, df2, on="key", how="right")
# In Pandas, a Right Join keeps all rows from the right DataFrame and matches rows from the left DataFrame.
# If a match is not found in the left table, the missing values become NaN.
```
