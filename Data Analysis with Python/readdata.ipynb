## Reading Data from Different Sources

### Reading the data from the json to df and again df to json

```python
import pandas as pd
import json
from io import StringIO

data = [
    {"Name": "Abhishek", "Age": 21, "City": "Delhi"},
    {"Name": "Rahul", "Age": 22, "City": "Mumbai"},
    {"Name": "Priya", "Age": 20, "City": "Pune"}
]

# Convert to JSON string
json_string = json.dumps(data)

# Convert string to DataFrame
df = pd.read_json(StringIO(json_string))

print(df)
```

```python
# Convert the df back into the json
json_output = df.to_json(orient="records")

print(json_output)
```

Orient :- In Pandas, the orient parameter in df.to_json() decides how the DataFrame is structured when converted to JSON. Different orient values change the format of the JSON output.

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Abhishek", "Rahul", "Priya"],
    "Age": [21, 22, 20],
    "City": ["Delhi", "Mumbai", "Pune"]
})
df
```

```python
# orient="records" (Most Common)
df.to_json(orient="records")
# Each row becomes a JSON object.
#Used in APIs and web applications.
```

```python
# orient="index" Here the index becomes the key.
df.to_json(orient="index")
```

```python
# orient="columns" (Default) Each column becomes the main key.
df.to_json()
```

## Read the data from an API or an URL

```python
import pandas as pd

# This is an URL who gives you the CSV file as the response

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

df = pd.read_csv(url)

print(df.head())
```

```python
# Sometimes an API returns CSV text instead of a file. Then you use requests.
import pandas as pd
import requests
from io import StringIO

url = "https://example.com/data.csv"

response = requests.get(url)

df = pd.read_csv(StringIO(response.text))

print(df.head())
```

```python
# API request
#    ↓
# CSV text response
#    ↓
# StringIO
#    ↓
# pd.read_csv()
#    ↓
# DataFrame
```

## Reading the HTML

In Pandas, read_html() is used to extract tables from an HTML webpage and convert them into DataFrames.
It automatically finds all 'table' tags in the HTML.


```python
# Basic Syntax : 
import pandas as pd
tables = pd.read_html("URL")

# 'tables' becomes a list of DataFrames
# Each table on the webpage becomes one DataFrame
```

```python
import pandas as pd

url = "https://www.w3schools.com/html/html_tables.asp"

tables = pd.read_html(url)

print(tables[0])  # This extracts the first table from the webpage.
```

```python
# Because read_html() returns a list of DataFrames.
# tables = [df1, df2, df3]
# To access them:

tables[0]   # first table
tables[1]   # second table
tables[2]   # third table
```

```python
# Save the Table as a DataFrame
df = pd.read_html(url)[0]
print(df.head())
```

## Read the data from the Pickle file
A Pickle file is a way to save Python objects to a file and load them later using pickle.
In data analysis, Pandas can store DataFrames as pickle files.
Pickle files usually have extensions like:
.pkl
.pickle


```python
# Save DataFrame as Pickle File
import pandas as pd

data = {
    "Name": ["Abhishek", "Rahul", "Priya"],
    "Age": [21, 22, 20]
}

df = pd.DataFrame(data)

df.to_pickle("data.pkl")
# This serializes the DataFrame and stores it in data.pkl.
```

```python
# Read Pickle File
import pandas as pd

df = pd.read_pickle("data.pkl")

print(df)
```

```python
# Using the pickle Module Directly
import pickle

# Save
with open("data.pkl", "wb") as f:
    pickle.dump(df, f)

# Load
with open("data.pkl", "rb") as f:
    df = pickle.load(f)
```
